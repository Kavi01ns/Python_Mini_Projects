## Pandas Series:  


```python
import pandas as pd
```

### 1. Creating Pandas Series:

```python
Mark = [95, 92, 89, 85, 80]
Marks = pd.Series(Mark, index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5'])
print(Marks)
```

### 2. Indexing and Slicing:

```python
Marks['Rank1']
```

```python
Marks.loc['Rank1':'Rank3']
```

```python
Marks.iloc[-3]
```

```python
Marks>90
```

```python
Marks[Marks>90]
```

### 3. Manipulating Series:

```python
Marks['Rank1']=100
Marks['Rank1']
```

```python
Marks = Marks / 10
Marks
```

```python
Marks = Marks.drop('Rank5')
Marks
```

## Pandas DataFrame:

### 1. Creating Pandas DataFrame:


```python
data = { 'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': ['Electronics', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
    'Region': ['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West', 'South', 'North'],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400] }

transactions = pd.DataFrame(data)
transactions
```

### 3. Manipulating the DataFrame:

```python
transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165
transactions
```

```python
transactions['Discount'] = transactions['Amount'] * 0.10
transactions
```

```python
transactions.drop(transactions[transactions['TransactionID'] == 109].index)
transactions
```

```python
transactions.drop(transactions[transactions['TransactionID'] == 109].index, axis=0)
transactions
```

```python
transactions.drop(columns=["Discount"],inplace=True)
transactions
```

### 4. Data Wrangling:

```python
df={'TransactionID': [101, 102, 103, 105, 106, 107, 108, 109, 110],
            'Profit': [50, 20, 70, 40, 60, 80, 30, 90, 110]}
profit_data = pd.DataFrame(df)
profit_data
```

```python
df_left = pd.merge(transactions, profit_data, on="TransactionID", how="left")
df_left
```

```python
df1= ({'TransactionID': [111, 112],'ProductCategory': ['Toys', 'Books'],
                               'Region': ['North', 'South'],'Amount': [220, 170] })
additional_data = pd.DataFrame(df1)
additional_data
```

```python
Concatenate = pd.concat([transactions, additional_data], ignore_index=True)
Concatenate
```
